```python
# TCAV Analysis for Slang Classification - Standalone Notebook
# This notebook runs the complete analysis and focuses on TCAV results

# =============================================================================
# FILE UPLOAD AND SETUP
# =============================================================================

from google.colab import files
import os

# Upload the data files
print("📁 Upload your data files:")
uploaded = files.upload()

# Install required packages
print("🔧 Installing packages...")
!pip install -q transformers scikit-learn pandas torch tqdm matplotlib seaborn

# Import all required libraries
import pandas as pd
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoModel
from sklearn.metrics import accuracy_score
from tqdm.auto import tqdm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# DATA LOADING (MINIMAL OUTPUT)
# =============================================================================

def load_data():
    """Load and prepare data silently"""
    df_slang = pd.read_csv("/content/slang_OpenSub_filtered.tsv", sep="\t", dtype=str)
    df_nonslang = pd.read_csv("/content/slang_OpenSub_negatives_filtered.tsv", sep="\t", dtype=str)

    df_slang["label"] = 1
    df_slang["confidence"] = df_slang["ANNOTATOR_CONFIDENCE"].astype(int)
    df_nonslang["label"] = 0
    df_nonslang["confidence"] = 0

    df_all = pd.concat([df_slang, df_nonslang], ignore_index=True)

    sentences = df_all["SENTENCE"].tolist()
    true_labels = df_all["label"].tolist()
    confidence_levels = df_all["confidence"].tolist()

    return sentences, true_labels, confidence_levels

# =============================================================================
# QUICK CLASSIFICATION (MINIMAL OUTPUT)
# =============================================================================

def quick_classify_model(model_name, sentences):
    """Run classification quickly without detailed output"""
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device).eval()

    predictions = []
    candidate_labels = ["slang", "non-slang"]

    for sentence in tqdm(sentences, desc=f"Classifying {model_name.split('/')[-1]}", leave=False):
        hypotheses = [f"This example is {label}." for label in candidate_labels]
        encoded = tokenizer(
            [sentence] * len(hypotheses),
            hypotheses,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=256
        ).to(device)

        with torch.no_grad():
            outputs = model(**encoded)
            logits = outputs.logits
            entail_logits = logits[:, 2]
            probs = F.softmax(entail_logits, dim=0)
            best_idx = torch.argmax(probs).item()

        predictions.append(1 if candidate_labels[best_idx] == "slang" else 0)

    return predictions

def run_all_classifications(sentences, true_labels):
    """Run all model classifications and return results"""
    models = {
        "BART-large-MNLI": "facebook/bart-large-mnli",
        "BERT-base-MNLI": "textattack/bert-base-uncased-MNLI",
        "RoBERTa-large-MNLI": "roberta-large-mnli",
        "DeBERTa-v3-large-MNLI": "MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli"
    }

    all_results = {}

    print("🚀 Running quick classifications...")
    for model_name, model_id in models.items():
        try:
            predictions = quick_classify_model(model_id, sentences)
            overall_acc = accuracy_score(true_labels, predictions)

            all_results[model_name] = {
                'model_id': model_id,
                'predictions': predictions,
                'accuracy': overall_acc
            }

            print(f"✅ {model_name}: {overall_acc:.4f}")

        except Exception as e:
            print(f"❌ {model_name}: Error - {str(e)}")

    return all_results

# =============================================================================
# ACTIVATION EXTRACTION
# =============================================================================

def get_model_activations(model_name, sentences, layer_idx=-2):
    """Extract hidden layer activations from a transformer model"""

    # Load model for feature extraction
    if "bart" in model_name.lower():
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModel.from_pretrained(model_name)
        encoder = model.encoder
    else:
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModel.from_pretrained(model_name)
        encoder = model

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    encoder.to(device).eval()

    activations = []

    for sentence in tqdm(sentences, desc=f"Extracting {model_name.split('/')[-1]} activations", leave=False):
        inputs = tokenizer(sentence, return_tensors="pt", padding=True,
                          truncation=True, max_length=128).to(device)

        with torch.no_grad():
            outputs = encoder(**inputs, output_hidden_states=True)
            hidden_states = outputs.hidden_states[layer_idx]
            pooled = torch.mean(hidden_states, dim=1).squeeze()
            activations.append(pooled.cpu().numpy())

    return np.array(activations)

# =============================================================================
# TCAV ANALYZER CLASS
# =============================================================================

class TCAV_Analyzer:
    def __init__(self, activations, labels, confidence_levels, model_predictions, model_name):
        self.activations = activations
        self.labels = labels
        self.confidence_levels = confidence_levels
        self.model_predictions = model_predictions
        self.model_name = model_name
        self.scaler = StandardScaler()
        self.activations_scaled = self.scaler.fit_transform(activations)

    def find_concept_direction(self):
        """Find the direction vector that separates slang from non-slang"""
        classifier = LogisticRegression(random_state=42, max_iter=1000)
        classifier.fit(self.activations_scaled, self.labels)

        concept_vector = classifier.coef_[0]
        concept_vector = concept_vector / np.linalg.norm(concept_vector)

        pred_labels = classifier.predict(self.activations_scaled)
        separation_accuracy = accuracy_score(self.labels, pred_labels)

        return concept_vector, classifier, separation_accuracy

    def compute_tcav_scores(self, concept_vector):
        """Compute TCAV scores for each example"""
        tcav_scores = np.dot(self.activations_scaled, concept_vector)
        return tcav_scores

    def analyze_bias_mathematically(self, tcav_scores):
        """Analyze bias mathematically"""
        slang_mask = np.array(self.labels) == 1
        nonslang_mask = np.array(self.labels) == 0

        slang_tcav_mean = np.mean(tcav_scores[slang_mask])
        nonslang_tcav_mean = np.mean(tcav_scores[nonslang_mask])

        # Bias metrics
        bias_strength = abs(slang_tcav_mean - nonslang_tcav_mean)
        bias_direction = "SLANG" if slang_tcav_mean > nonslang_tcav_mean else "NON-SLANG"

        # Misclassification analysis
        misclassified_mask = (self.labels != self.model_predictions)
        slang_misclassified = misclassified_mask & (self.labels == 1)
        nonslang_misclassified = misclassified_mask & (self.labels == 0)

        # Statistical test for bias
        from scipy import stats
        slang_scores = tcav_scores[slang_mask]
        nonslang_scores = tcav_scores[nonslang_mask]
        t_stat, p_value = stats.ttest_ind(slang_scores, nonslang_scores)

        results = {
            'slang_tcav_mean': slang_tcav_mean,
            'nonslang_tcav_mean': nonslang_tcav_mean,
            'bias_strength': bias_strength,
            'bias_direction': bias_direction,
            'slang_misclassified_tcav': tcav_scores[slang_misclassified],
            'nonslang_misclassified_tcav': tcav_scores[nonslang_misclassified],
            't_statistic': t_stat,
            'p_value': p_value,
            'effect_size': bias_strength / np.std(tcav_scores)
        }

        return results

    def analyze_confidence_correlation(self, tcav_scores):
        """Analyze correlation between confidence and TCAV scores"""
        confidence_analysis = {}

        for conf_level in [1, 2, 3]:
            mask = (np.array(self.confidence_levels) == conf_level) & (np.array(self.labels) == 1)
            if np.sum(mask) > 0:
                conf_tcav_scores = tcav_scores[mask]
                conf_predictions = np.array(self.model_predictions)[mask]
                conf_labels = np.array(self.labels)[mask]

                confidence_analysis[conf_level] = {
                    'mean_tcav': np.mean(conf_tcav_scores),
                    'std_tcav': np.std(conf_tcav_scores),
                    'accuracy': accuracy_score(conf_labels, conf_predictions),
                    'count': len(conf_tcav_scores)
                }

        return confidence_analysis

    def create_tcav_visualizations(self, concept_vector, tcav_scores, bias_analysis):
        """Create comprehensive TCAV visualizations"""
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle(f'TCAV Analysis: {self.model_name}', fontsize=16, fontweight='bold')

        # 1. PCA with concept direction
        pca = PCA(n_components=2)
        activations_2d = pca.fit_transform(self.activations_scaled)
        concept_2d = pca.transform(concept_vector.reshape(1, -1))[0]
        concept_2d = concept_2d / np.linalg.norm(concept_2d) * 3

        ax1 = axes[0, 0]
        scatter = ax1.scatter(activations_2d[:, 0], activations_2d[:, 1],
                             c=self.labels, cmap='RdYlBu', alpha=0.6, s=20)
        ax1.arrow(0, 0, concept_2d[0], concept_2d[1], head_width=0.15,
                  head_length=0.15, fc='black', ec='black', linewidth=3)
        ax1.set_title('Concept Direction in PCA Space')
        ax1.set_xlabel('PC1')
        ax1.set_ylabel('PC2')
        ax1.grid(True, alpha=0.3)

        # 2. TCAV scores distribution
        ax2 = axes[0, 1]
        slang_mask = np.array(self.labels) == 1
        nonslang_mask = np.array(self.labels) == 0

        ax2.hist(tcav_scores[slang_mask], bins=40, alpha=0.7, label='Slang', color='red', density=True)
        ax2.hist(tcav_scores[nonslang_mask], bins=40, alpha=0.7, label='Non-Slang', color='blue', density=True)
        ax2.axvline(0, color='black', linestyle='--', linewidth=2, label='Neutral')
        ax2.axvline(bias_analysis['slang_tcav_mean'], color='red', linestyle='-', alpha=0.8, label='Slang Mean')
        ax2.axvline(bias_analysis['nonslang_tcav_mean'], color='blue', linestyle='-', alpha=0.8, label='Non-Slang Mean')
        ax2.set_title('TCAV Scores Distribution')
        ax2.set_xlabel('TCAV Score')
        ax2.set_ylabel('Density')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        # 3. Misclassification analysis
        ax3 = axes[0, 2]
        correct_mask = np.array(self.labels) == np.array(self.model_predictions)

        ax3.scatter(tcav_scores[correct_mask & slang_mask], [1]*np.sum(correct_mask & slang_mask),
                   alpha=0.6, label='Slang Correct', color='darkgreen', s=15)
        ax3.scatter(tcav_scores[~correct_mask & slang_mask], [1]*np.sum(~correct_mask & slang_mask),
                   alpha=0.8, label='Slang Wrong', color='red', marker='x', s=30)
        ax3.scatter(tcav_scores[correct_mask & nonslang_mask], [0]*np.sum(correct_mask & nonslang_mask),
                   alpha=0.6, label='Non-Slang Correct', color='darkblue', s=15)
        ax3.scatter(tcav_scores[~correct_mask & nonslang_mask], [0]*np.sum(~correct_mask & nonslang_mask),
                   alpha=0.8, label='Non-Slang Wrong', color='orange', marker='x', s=30)
        ax3.axvline(0, color='black', linestyle='--', alpha=0.7)
        ax3.set_title('Misclassifications in TCAV Space')
        ax3.set_xlabel('TCAV Score')
        ax3.set_ylabel('True Label')
        ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax3.grid(True, alpha=0.3)

        # 4. Confidence analysis
        ax4 = axes[1, 0]
        colors = ['lightcoral', 'orange', 'darkred']
        for i, conf_level in enumerate([1, 2, 3]):
            mask = (np.array(self.confidence_levels) == conf_level) & (np.array(self.labels) == 1)
            if np.sum(mask) > 0:
                ax4.scatter(tcav_scores[mask], [conf_level] * np.sum(mask),
                           alpha=0.7, color=colors[i], label=f'Confidence {conf_level}', s=20)
        ax4.axvline(0, color='black', linestyle='--', alpha=0.7)
        ax4.set_title('Confidence Levels vs TCAV Scores')
        ax4.set_xlabel('TCAV Score')
        ax4.set_ylabel('Confidence Level')
        ax4.legend()
        ax4.grid(True, alpha=0.3)

        # 5. 3D PCA visualization
        ax5 = axes[1, 1]
        ax5.remove()
        ax5 = fig.add_subplot(2, 3, 5, projection='3d')

        pca_3d = PCA(n_components=3)
        activations_3d = pca_3d.fit_transform(self.activations_scaled)
        concept_3d = pca_3d.transform(concept_vector.reshape(1, -1))[0]
        concept_3d = concept_3d / np.linalg.norm(concept_3d) * 3

        ax5.scatter(activations_3d[slang_mask, 0], activations_3d[slang_mask, 1], activations_3d[slang_mask, 2],
                   c='red', alpha=0.6, label='Slang', s=15)
        ax5.scatter(activations_3d[nonslang_mask, 0], activations_3d[nonslang_mask, 1], activations_3d[nonslang_mask, 2],
                   c='blue', alpha=0.6, label='Non-Slang', s=15)
        ax5.quiver(0, 0, 0, concept_3d[0], concept_3d[1], concept_3d[2],
                  color='black', arrow_length_ratio=0.1, linewidth=3)
        ax5.set_title('3D PCA with Concept Vector')
        ax5.legend()

        # 6. Bias strength visualization
        ax6 = axes[1, 2]

        # Create bias strength bar
        bias_data = [bias_analysis['slang_tcav_mean'], bias_analysis['nonslang_tcav_mean']]
        colors = ['red', 'blue']
        bars = ax6.bar(['Slang\nMean', 'Non-Slang\nMean'], bias_data, color=colors, alpha=0.7)
        ax6.axhline(0, color='black', linestyle='-', linewidth=1)
        ax6.set_title(f"Bias Strength: {bias_analysis['bias_strength']:.3f}")
        ax6.set_ylabel('Mean TCAV Score')
        ax6.grid(True, alpha=0.3)

        # Add value labels on bars
        for bar, value in zip(bars, bias_data):
            height = bar.get_height()
            ax6.text(bar.get_x() + bar.get_width()/2., height + (0.01 if height >= 0 else -0.03),
                    f'{value:.3f}', ha='center', va='bottom' if height >= 0 else 'top')

        plt.tight_layout()

        # Save figure
        filename = f'/content/tcav_analysis_{self.model_name.replace("-", "_")}.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')

        return fig, filename

# =============================================================================
# MAIN TCAV EXECUTION
# =============================================================================

def run_complete_tcav_analysis():
    """Run complete TCAV analysis pipeline"""
    print("🚀 Starting Complete TCAV Analysis Pipeline")
    print("=" * 60)

    # Load data
    print("📂 Loading data...")
    sentences, true_labels, confidence_levels = load_data()
    print(f"✅ Loaded {len(sentences)} sentences")

    # Run classifications
    print("\n🔍 Running model classifications...")
    all_results = run_all_classifications(sentences, true_labels)

    # TCAV Analysis
    print(f"\n🧠 Starting TCAV Analysis...")

    tcav_results = {}

    for model_name, result in all_results.items():
        print(f"\n{'='*50}")
        print(f"🔬 TCAV Analysis: {model_name}")
        print(f"{'='*50}")

        try:
            # Extract activations
            print("📊 Extracting activations...")
            activations = get_model_activations(result['model_id'], sentences)

            # Create TCAV analyzer
            tcav_analyzer = TCAV_Analyzer(
                activations=activations,
                labels=true_labels,
                confidence_levels=confidence_levels,
                model_predictions=result['predictions'],
                model_name=model_name
            )

            # Find concept direction
            print("🎯 Finding concept direction...")
            concept_vector, classifier, separation_accuracy = tcav_analyzer.find_concept_direction()

            # Compute TCAV scores
            print("📈 Computing TCAV scores...")
            tcav_scores = tcav_analyzer.compute_tcav_scores(concept_vector)

            # Mathematical bias analysis
            print("🔢 Analyzing bias mathematically...")
            bias_analysis = tcav_analyzer.analyze_bias_mathematically(tcav_scores)

            # Confidence correlation
            print("🎭 Analyzing confidence correlation...")
            confidence_analysis = tcav_analyzer.analyze_confidence_correlation(tcav_scores)

            # Create visualizations
            print("📊 Creating visualizations...")
            fig, filename = tcav_analyzer.create_tcav_visualizations(concept_vector, tcav_scores, bias_analysis)

            # Store results
            tcav_results[model_name] = {
                'concept_vector': concept_vector,
                'tcav_scores': tcav_scores,
                'separation_accuracy': separation_accuracy,
                'bias_analysis': bias_analysis,
                'confidence_analysis': confidence_analysis,
                'visualization_file': filename
            }

            # Print key results
            print(f"\n📊 KEY RESULTS:")
            print(f"   Separation Accuracy: {separation_accuracy:.4f}")
            print(f"   Bias Direction: {bias_analysis['bias_direction']}")
            print(f"   Bias Strength: {bias_analysis['bias_strength']:.4f}")
            print(f"   Statistical Significance: p={bias_analysis['p_value']:.2e}")
            print(f"   Effect Size: {bias_analysis['effect_size']:.3f}")
            print(f"   Visualization saved: {filename}")

        except Exception as e:
            print(f"❌ Error in TCAV analysis for {model_name}: {str(e)}")
            continue

    # Final comparative analysis
    print(f"\n{'='*80}")
    print(f"📊 FINAL TCAV COMPARISON")
    print(f"{'='*80}")

    if tcav_results:
        # Create summary table
        print(f"{'Model':<25} {'Sep.Acc':<8} {'Bias.Dir':<10} {'Bias.Str':<8} {'Effect.Size':<10} {'P-Value':<10}")
        print("-" * 80)

        for model_name, results in tcav_results.items():
            bias = results['bias_analysis']
            print(f"{model_name:<25} {results['separation_accuracy']:<8.3f} {bias['bias_direction']:<10} "
                  f"{bias['bias_strength']:<8.3f} {bias['effect_size']:<10.3f} {bias['p_value']:<10.2e}")

        # Mathematical proof summary
        print(f"\n🔬 MATHEMATICAL PROOF OF BIAS:")
        for model_name, results in tcav_results.items():
            bias = results['bias_analysis']
            print(f"\n   {model_name}:")
            print(f"      • Slang TCAV Mean: {bias['slang_tcav_mean']:.3f}")
            print(f"      • Non-Slang TCAV Mean: {bias['nonslang_tcav_mean']:.3f}")
            print(f"      • Difference: {bias['slang_tcav_mean'] - bias['nonslang_tcav_mean']:.3f}")
            print(f"      • Effect Size: {bias['effect_size']:.3f} (Cohen's d)")
            print(f"      • Statistical Test: t={bias['t_statistic']:.3f}, p={bias['p_value']:.2e}")

            if bias['p_value'] < 0.001:
                print(f"      ✅ HIGHLY SIGNIFICANT BIAS toward {bias['bias_direction']}")
            elif bias['p_value'] < 0.05:
                print(f"      ✅ SIGNIFICANT BIAS toward {bias['bias_direction']}")
            else:
                print(f"      ❌ No significant bias detected")

        # Save results
        print(f"\n💾 Saving comprehensive results...")
        np.save('/content/complete_tcav_results.npy', tcav_results)

        # Create summary CSV
        summary_data = []
        for model_name, results in tcav_results.items():
            bias = results['bias_analysis']
            summary_data.append({
                'model': model_name,
                'separation_accuracy': results['separation_accuracy'],
                'bias_direction': bias['bias_direction'],
                'bias_strength': bias['bias_strength'],
                'slang_tcav_mean': bias['slang_tcav_mean'],
                'nonslang_tcav_mean': bias['nonslang_tcav_mean'],
                'effect_size': bias['effect_size'],
                'p_value': bias['p_value'],
                't_statistic': bias['t_statistic']
            })

        summary_df = pd.DataFrame(summary_data)
        summary_df.to_csv('/content/tcav_bias_summary.csv', index=False)

        print(f"✅ Results saved:")
        print(f"   📁 /content/complete_tcav_results.npy")
        print(f"   📁 /content/tcav_bias_summary.csv")
        print(f"   📁 Individual visualization files")

    print(f"\n🎉 TCAV Analysis Complete!")
    return tcav_results

# =============================================================================
# RUN ANALYSIS
# =============================================================================

if __name__ == "__main__":
    # Install scipy for statistical tests
    !pip install -q scipy
    from scipy import stats

    # Run the complete analysis
    tcav_results = run_complete_tcav_analysis()
```
