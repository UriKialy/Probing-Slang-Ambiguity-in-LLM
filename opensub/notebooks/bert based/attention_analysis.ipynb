```python
# Attention Analysis for Slang Detection - Slang Term Focused Analysis
# This notebook analyzes which attention heads specifically focus on slang terms

# =============================================================================
# FILE UPLOAD AND SETUP
# =============================================================================

from google.colab import files
import os

# Upload the data files
print("📁 Upload your data files:")
uploaded = files.upload()

# Install required packages
print("🔧 Installing packages...")
!pip install -q transformers scikit-learn pandas torch tqdm matplotlib seaborn plotly bertviz

# Import all required libraries
import pandas as pd
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModel, AutoModelForSequenceClassification
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tqdm.auto import tqdm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import warnings
import re
warnings.filterwarnings('ignore')

# =============================================================================
# DATA LOADING AND SLANG TERM EXTRACTION
# =============================================================================

def load_and_prepare_slang_data(max_samples_per_class=250):
    """Load data and extract slang terms for focused analysis"""
    print("📂 Loading data files...")

    try:
        df_slang = pd.read_csv("/content/slang_OpenSub_filtered.tsv", sep="\t", dtype=str)
        df_nonslang = pd.read_csv("/content/slang_OpenSub_negatives_filtered.tsv", sep="\t", dtype=str)

        print(f"   Loaded {len(df_slang)} slang sentences")
        print(f"   Loaded {len(df_nonslang)} non-slang sentences")

    except FileNotFoundError as e:
        print(f"❌ File not found: {e}")
        print("Please ensure both TSV files are uploaded to Colab")
        return None

    # Prepare slang data with slang terms
    df_slang = df_slang.dropna(subset=['SENTENCE', 'SLANG_TERM'])
    df_slang["label"] = 1
    df_slang["confidence"] = df_slang["ANNOTATOR_CONFIDENCE"].astype(int) if "ANNOTATOR_CONFIDENCE" in df_slang.columns else 1

    # Prepare non-slang data
    df_nonslang = df_nonslang.dropna(subset=['SENTENCE'])
    df_nonslang["label"] = 0
    df_nonslang["confidence"] = 0
    df_nonslang["SLANG_TERM"] = ""  # No slang terms in non-slang data

    # Balance the dataset
    n_slang = min(len(df_slang), max_samples_per_class)
    n_nonslang = min(len(df_nonslang), max_samples_per_class)

    # Sample equal amounts from each class
    df_slang_sampled = df_slang.sample(n=n_slang, random_state=42)
    df_nonslang_sampled = df_nonslang.sample(n=n_nonslang, random_state=42)

    # Combine datasets
    df_all = pd.concat([df_slang_sampled, df_nonslang_sampled], ignore_index=True)
    df_all = df_all.sample(frac=1, random_state=42).reset_index(drop=True)

    # Extract unique slang terms for analysis
    unique_slang_terms = df_slang_sampled['SLANG_TERM'].unique()
    slang_term_counts = df_slang_sampled['SLANG_TERM'].value_counts()

    print(f"✅ Prepared dataset with slang term analysis:")
    print(f"   Total sentences: {len(df_all)}")
    print(f"   Slang samples: {sum(df_all['label'])}")
    print(f"   Non-slang samples: {len(df_all) - sum(df_all['label'])}")
    print(f"   Unique slang terms: {len(unique_slang_terms)}")
    print(f"   Most common slang terms: {list(slang_term_counts.head().index)}")

    return df_all, unique_slang_terms, slang_term_counts

# =============================================================================
# SLANG-FOCUSED ATTENTION ANALYZER
# =============================================================================

class SlangAttentionAnalyzer:
    def __init__(self, model_name, model_id):
        self.model_name = model_name
        self.model_id = model_id
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

        print(f"🔧 Loading {model_name}...")

        try:
            # Load model and tokenizer
            self.tokenizer = AutoTokenizer.from_pretrained(model_id)
            self.model = AutoModel.from_pretrained(model_id, output_attentions=True)
            self.model.to(self.device).eval()

            # Get model architecture info
            self.num_layers = self.model.config.num_hidden_layers
            self.num_heads = self.model.config.num_attention_heads

            print(f"✅ Loaded {model_name}: {self.num_layers} layers, {self.num_heads} heads")

        except Exception as e:
            print(f"❌ Error loading {model_name}: {str(e)}")
            raise

    def find_slang_tokens(self, tokens, slang_term):
        """Find the position of slang term tokens in the tokenized sequence"""
        if not slang_term:
            return []

        # Clean and normalize slang term
        slang_term_clean = slang_term.lower().strip()

        # Try different tokenization approaches
        slang_tokens = self.tokenizer.tokenize(slang_term_clean)

        # Find positions in the token sequence
        slang_positions = []

        # Direct substring matching
        for i, token in enumerate(tokens):
            token_clean = token.lower().replace('##', '').replace('Ġ', '').replace('▁', '')
            if slang_term_clean in token_clean or token_clean in slang_term_clean:
                slang_positions.append(i)

        # Multi-token slang terms
        if len(slang_tokens) > 1:
            for i in range(len(tokens) - len(slang_tokens) + 1):
                sequence_match = True
                for j, slang_token in enumerate(slang_tokens):
                    token_clean = tokens[i + j].lower().replace('##', '').replace('Ġ', '').replace('▁', '')
                    slang_token_clean = slang_token.lower().replace('##', '').replace('Ġ', '').replace('▁', '')
                    if slang_token_clean not in token_clean and token_clean not in slang_token_clean:
                        sequence_match = False
                        break
                if sequence_match:
                    slang_positions.extend(range(i, i + len(slang_tokens)))

        return list(set(slang_positions))  # Remove duplicates

    def extract_slang_attention_patterns(self, df_data, max_samples=300):
        """Extract attention patterns with focus on slang terms"""
        print(f"📊 Extracting slang-focused attention patterns...")

        # Ensure balanced sampling but prioritize examples with clear slang terms
        slang_data = df_data[df_data['label'] == 1].copy()
        nonslang_data = df_data[df_data['label'] == 0].copy()

        # Prioritize slang examples with shorter, clearer slang terms
        slang_data['slang_len'] = slang_data['SLANG_TERM'].str.len()
        slang_data = slang_data.sort_values('slang_len')  # Shorter terms first

        n_per_class = min(max_samples // 2, len(slang_data), len(nonslang_data))

        selected_slang = slang_data.head(n_per_class)
        selected_nonslang = nonslang_data.sample(n=n_per_class, random_state=42)

        selected_data = pd.concat([selected_slang, selected_nonslang]).sample(frac=1, random_state=42)

        print(f"   Processing {len(selected_data)} sentences")
        print(f"   Slang examples: {sum(selected_data['label'])}")
        print(f"   Non-slang examples: {len(selected_data) - sum(selected_data['label'])}")

        attention_data = []
        slang_attention_stats = {}

        for idx, row in tqdm(selected_data.iterrows(), desc=f"Processing {self.model_name}", total=len(selected_data)):
            try:
                sentence = row['SENTENCE']
                label = row['label']
                slang_term = row['SLANG_TERM'] if pd.notna(row['SLANG_TERM']) else ""

                if not sentence or len(sentence.strip()) < 3:
                    continue

                # Tokenize
                inputs = self.tokenizer(sentence, return_tensors="pt", padding=True,
                                      truncation=True, max_length=128).to(self.device)

                # Get tokens for analysis
                tokens = self.tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])

                # Find slang term positions
                slang_positions = self.find_slang_tokens(tokens, slang_term) if slang_term else []

                with torch.no_grad():
                    outputs = self.model(**inputs, output_attentions=True)

                    # Process each layer's attention
                    for layer_idx, layer_attention in enumerate(outputs.attentions):
                        layer_attention = layer_attention.squeeze(0).cpu().numpy()

                        # Analyze each attention head
                        for head_idx in range(layer_attention.shape[0]):
                            head_attention = layer_attention[head_idx]

                            # Analyze slang-specific attention
                            slang_analysis = self.analyze_slang_attention(
                                head_attention, tokens, slang_positions, slang_term,
                                layer_idx, head_idx
                            )

                            attention_data.append({
                                'sentence_idx': idx,
                                'sentence': sentence,
                                'label': label,
                                'slang_term': slang_term,
                                'layer': layer_idx,
                                'head': head_idx,
                                'tokens': tokens,
                                'slang_positions': slang_positions,
                                'attention_matrix': head_attention,
                                **slang_analysis
                            })

                            # Collect slang-specific statistics
                            if slang_term and slang_positions:
                                key = (layer_idx, head_idx)
                                if key not in slang_attention_stats:
                                    slang_attention_stats[key] = {
                                        'slang_focus_scores': [],
                                        'slang_terms_seen': set(),
                                        'examples_count': 0
                                    }

                                slang_attention_stats[key]['slang_focus_scores'].append(
                                    slang_analysis['slang_focus_score']
                                )
                                slang_attention_stats[key]['slang_terms_seen'].add(slang_term)
                                slang_attention_stats[key]['examples_count'] += 1

            except Exception as e:
                print(f"⚠️ Error processing sentence {idx}: {str(e)}")
                continue

        print(f"✅ Extracted attention patterns from {len(attention_data)} head instances")
        print(f"✅ Collected slang statistics for {len(slang_attention_stats)} head combinations")

        return attention_data, slang_attention_stats

    def analyze_slang_attention(self, attention_matrix, tokens, slang_positions, slang_term, layer_idx, head_idx):
        """Analyze how much attention is focused on slang terms"""
        seq_len = len(tokens)
        analysis = {}

        if not slang_positions or len(slang_positions) == 0:
            # No slang term found
            analysis.update({
                'slang_focus_score': 0.0,
                'to_slang_attention': 0.0,
                'from_slang_attention': 0.0,
                'slang_self_attention': 0.0,
                'context_to_slang': 0.0,
                'slang_to_context': 0.0,
                'slang_centrality': 0.0,
                'has_slang_term': False
            })
            return analysis

        # 1. Overall slang focus score (how much attention goes TO slang tokens)
        to_slang_scores = []
        for i in range(seq_len):
            for slang_pos in slang_positions:
                if i != slang_pos:  # Don't include self-attention
                    to_slang_scores.append(attention_matrix[i, slang_pos])

        analysis['to_slang_attention'] = np.mean(to_slang_scores) if to_slang_scores else 0.0

        # 2. Attention FROM slang tokens to other tokens
        from_slang_scores = []
        for slang_pos in slang_positions:
            for j in range(seq_len):
                if j != slang_pos:
                    from_slang_scores.append(attention_matrix[slang_pos, j])

        analysis['from_slang_attention'] = np.mean(from_slang_scores) if from_slang_scores else 0.0

        # 3. Self-attention of slang tokens
        slang_self_scores = [attention_matrix[pos, pos] for pos in slang_positions]
        analysis['slang_self_attention'] = np.mean(slang_self_scores) if slang_self_scores else 0.0

        # 4. Context tokens attending to slang (excluding immediate neighbors)
        context_to_slang_scores = []
        for i in range(seq_len):
            if i not in slang_positions:
                # Check if this is not an immediate neighbor
                is_neighbor = any(abs(i - pos) <= 1 for pos in slang_positions)
                if not is_neighbor:
                    for slang_pos in slang_positions:
                        context_to_slang_scores.append(attention_matrix[i, slang_pos])

        analysis['context_to_slang'] = np.mean(context_to_slang_scores) if context_to_slang_scores else 0.0

        # 5. Slang tokens attending to distant context
        slang_to_context_scores = []
        for slang_pos in slang_positions:
            for j in range(seq_len):
                if j not in slang_positions and abs(j - slang_pos) > 1:
                    slang_to_context_scores.append(attention_matrix[slang_pos, j])

        analysis['slang_to_context'] = np.mean(slang_to_context_scores) if slang_to_context_scores else 0.0

        # 6. Slang centrality (how much slang is a hub of attention)
        centrality_scores = []
        for slang_pos in slang_positions:
            # Incoming attention (excluding self)
            incoming = sum(attention_matrix[i, slang_pos] for i in range(seq_len) if i != slang_pos)
            # Outgoing attention (excluding self)
            outgoing = sum(attention_matrix[slang_pos, j] for j in range(seq_len) if j != slang_pos)
            centrality_scores.append(incoming + outgoing)

        analysis['slang_centrality'] = np.mean(centrality_scores) if centrality_scores else 0.0

        # 7. Overall slang focus score (composite metric)
        analysis['slang_focus_score'] = (
            analysis['to_slang_attention'] * 0.4 +
            analysis['context_to_slang'] * 0.3 +
            analysis['slang_centrality'] * 0.2 +
            analysis['slang_self_attention'] * 0.1
        )

        analysis['has_slang_term'] = True

        return analysis

    def identify_slang_specialist_heads(self, attention_data, slang_attention_stats):
        """Identify attention heads that specialize in processing slang terms"""
        print("🎯 Identifying slang specialist attention heads...")

        # Compute head-level slang specialization scores
        head_specializations = []

        for (layer, head), stats in slang_attention_stats.items():
            if stats['examples_count'] >= 5:  # Minimum examples for reliable statistics
                avg_slang_focus = np.mean(stats['slang_focus_scores'])
                std_slang_focus = np.std(stats['slang_focus_scores'])
                unique_terms = len(stats['slang_terms_seen'])

                # Compute relative specialization by comparing to non-slang performance
                non_slang_scores = []
                for data in attention_data:
                    if (data['layer'] == layer and data['head'] == head and
                        data['label'] == 0):  # Non-slang examples
                        non_slang_scores.append(data['slang_focus_score'])

                avg_non_slang_focus = np.mean(non_slang_scores) if non_slang_scores else 0.0

                specialization_score = avg_slang_focus - avg_non_slang_focus

                head_specializations.append({
                    'layer': layer,
                    'head': head,
                    'avg_slang_focus': avg_slang_focus,
                    'std_slang_focus': std_slang_focus,
                    'avg_non_slang_focus': avg_non_slang_focus,
                    'specialization_score': specialization_score,
                    'unique_slang_terms': unique_terms,
                    'examples_count': stats['examples_count'],
                    'consistency': 1.0 / (std_slang_focus + 0.01)  # Higher is more consistent
                })

        # Sort by specialization score
        head_specializations.sort(key=lambda x: x['specialization_score'], reverse=True)

        return head_specializations

    def visualize_slang_specialists(self, head_specializations, slang_terms_counts):
        """Create visualizations of slang specialist heads"""
        print("🎨 Creating slang specialist visualizations...")

        if not head_specializations:
            print("⚠️ No head specializations to visualize")
            return None

        df = pd.DataFrame(head_specializations)

        # Create comprehensive figure
        fig, axes = plt.subplots(3, 3, figsize=(20, 18))
        fig.suptitle(f'Slang Specialist Attention Heads: {self.model_name}', fontsize=16, fontweight='bold')

        # 1. Top slang specialists
        ax1 = axes[0, 0]
        top_specialists = df.head(15)
        y_pos = range(len(top_specialists))
        bars = ax1.barh(y_pos, top_specialists['specialization_score'],
                       color='darkblue', alpha=0.7)
        ax1.set_yticks(y_pos)
        ax1.set_yticklabels([f"L{int(row['layer'])}-H{int(row['head'])}"
                           for _, row in top_specialists.iterrows()])
        ax1.set_title('Top Slang Specialist Heads')
        ax1.set_xlabel('Slang Specialization Score')
        ax1.grid(True, alpha=0.3)

        # Add value labels on bars
        for i, bar in enumerate(bars):
            width = bar.get_width()
            ax1.text(width, bar.get_y() + bar.get_height()/2,
                    f'{width:.3f}', ha='left', va='center', fontsize=8)

        # 2. Specialization by layer
        ax2 = axes[0, 1]
        layer_specialization = df.groupby('layer')['specialization_score'].mean()
        ax2.plot(layer_specialization.index, layer_specialization.values, 'ro-', linewidth=2)
        ax2.set_title('Average Slang Specialization by Layer')
        ax2.set_xlabel('Layer')
        ax2.set_ylabel('Average Specialization Score')
        ax2.grid(True, alpha=0.3)

        # 3. Specialization heatmap
        ax3 = axes[0, 2]
        specialization_matrix = np.zeros((self.num_layers, self.num_heads))
        for _, row in df.iterrows():
            layer, head = int(row['layer']), int(row['head'])
            if layer < self.num_layers and head < self.num_heads:
                specialization_matrix[layer, head] = row['specialization_score']

        im3 = ax3.imshow(specialization_matrix, cmap='RdYlBu_r', aspect='auto')
        ax3.set_title('Slang Specialization Heatmap')
        ax3.set_xlabel('Head')
        ax3.set_ylabel('Layer')
        ax3.set_xticks(range(0, self.num_heads, 2))
        ax3.set_yticks(range(self.num_layers))
        plt.colorbar(im3, ax=ax3, label='Specialization Score')

        # 4. Consistency vs Specialization
        ax4 = axes[1, 0]
        scatter = ax4.scatter(df['consistency'], df['specialization_score'],
                            c=df['layer'], cmap='viridis', alpha=0.6, s=60)
        ax4.set_xlabel('Consistency (1/std)')
        ax4.set_ylabel('Specialization Score')
        ax4.set_title('Consistency vs Specialization')
        plt.colorbar(scatter, ax=ax4, label='Layer')

        # 5. Examples processed vs specialization
        ax5 = axes[1, 1]
        ax5.scatter(df['examples_count'], df['specialization_score'], alpha=0.6)
        ax5.set_xlabel('Number of Examples Processed')
        ax5.set_ylabel('Specialization Score')
        ax5.set_title('Examples vs Specialization')
        ax5.grid(True, alpha=0.3)

        # 6. Distribution of specialization scores
        ax6 = axes[1, 2]
        ax6.hist(df['specialization_score'], bins=20, alpha=0.7, edgecolor='black')
        ax6.axvline(df['specialization_score'].mean(), color='red', linestyle='--',
                   label=f'Mean: {df["specialization_score"].mean():.3f}')
        ax6.set_title('Distribution of Specialization Scores')
        ax6.set_xlabel('Specialization Score')
        ax6.set_ylabel('Number of Heads')
        ax6.legend()

        # 7. Top slang terms processed
        ax7 = axes[2, 0]
        top_terms = slang_terms_counts.head(10)
        ax7.barh(range(len(top_terms)), top_terms.values)
        ax7.set_yticks(range(len(top_terms)))
        ax7.set_yticklabels(top_terms.index)
        ax7.set_title('Most Common Slang Terms in Dataset')
        ax7.set_xlabel('Frequency')

        # 8. Slang vs Non-slang focus comparison
        ax8 = axes[2, 1]
        comparison_data = df[['avg_slang_focus', 'avg_non_slang_focus']].head(10)
        x = range(len(comparison_data))
        width = 0.35
        ax8.bar([i - width/2 for i in x], comparison_data['avg_slang_focus'],
               width, label='Slang Focus', alpha=0.7)
        ax8.bar([i + width/2 for i in x], comparison_data['avg_non_slang_focus'],
               width, label='Non-slang Focus', alpha=0.7)
        ax8.set_xlabel('Top Specialist Heads')
        ax8.set_ylabel('Average Focus Score')
        ax8.set_title('Slang vs Non-slang Focus (Top 10)')
        ax8.set_xticks(x)
        ax8.set_xticklabels([f"L{int(row['layer'])}-H{int(row['head'])}"
                           for _, row in comparison_data.iterrows()], rotation=45)
        ax8.legend()

        # 9. Unique terms coverage
        ax9 = axes[2, 2]
        ax9.scatter(df['unique_slang_terms'], df['specialization_score'], alpha=0.6)
        ax9.set_xlabel('Unique Slang Terms Seen')
        ax9.set_ylabel('Specialization Score')
        ax9.set_title('Term Diversity vs Specialization')
        ax9.grid(True, alpha=0.3)

        plt.tight_layout()

        # Save figure
        filename = f'/content/slang_specialists_{self.model_name.replace("-", "_").replace("/", "_")}.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        plt.show()

        return filename

    def create_slang_analysis_summary(self, head_specializations, slang_terms_counts):
        """Create detailed summary of slang analysis"""
        if not head_specializations:
            print("⚠️ No specializations found")
            return None

        df = pd.DataFrame(head_specializations)

        print(f"\n📊 SLANG SPECIALIST ANALYSIS for {self.model_name}")
        print("=" * 70)

        # Top specialists
        print(f"\n🏆 TOP 10 SLANG SPECIALIST HEADS:")
        top_10 = df.head(10)
        for i, (_, head) in enumerate(top_10.iterrows(), 1):
            layer, head_num = int(head['layer']), int(head['head'])
            score = head['specialization_score']
            consistency = head['consistency']
            terms = head['unique_slang_terms']
            print(f"   {i:2d}. Layer {layer:2d}, Head {head_num:2d}: "
                  f"Score={score:6.4f}, Consistency={consistency:5.2f}, Terms={terms:2d}")

        # Layer analysis
        print(f"\n📊 LAYER-WISE ANALYSIS:")
        layer_stats = df.groupby('layer').agg({
            'specialization_score': ['mean', 'max', 'count'],
            'unique_slang_terms': 'mean'
        }).round(4)

        print(f"   {'Layer':<6} {'Avg Score':<12} {'Max Score':<12} {'Heads':<8} {'Avg Terms':<12}")
        print("   " + "-" * 55)
        for layer in range(self.num_layers):
            if layer in layer_stats.index:
                stats = layer_stats.loc[layer]
                avg_score = stats[('specialization_score', 'mean')]
                max_score = stats[('specialization_score', 'max')]
                head_count = int(stats[('specialization_score', 'count')])
                avg_terms = stats[('unique_slang_terms', 'mean')]
                print(f"   {layer:<6} {avg_score:<12.4f} {max_score:<12.4f} {head_count:<8} {avg_terms:<12.2f}")

        # Best performing layer
        best_layer = df.loc[df['specialization_score'].idxmax(), 'layer']
        best_score = df['specialization_score'].max()
        print(f"\n🎯 BEST PERFORMING LAYER: {int(best_layer)} (Score: {best_score:.4f})")

        # Specialization statistics
        positive_specialists = len(df[df['specialization_score'] > 0])
        total_heads = len(df)
        print(f"\n📈 SPECIALIZATION STATISTICS:")
        print(f"   Heads with positive specialization: {positive_specialists}/{total_heads} ({100*positive_specialists/total_heads:.1f}%)")
        print(f"   Average specialization score: {df['specialization_score'].mean():.4f}")
        print(f"   Standard deviation: {df['specialization_score'].std():.4f}")

        return df

# =============================================================================
# MAIN ANALYSIS FUNCTION
# =============================================================================

def run_slang_specialist_analysis():
    """Run complete slang specialist analysis for all models"""
    print("🚀 Starting Slang Specialist Attention Analysis")
    print("=" * 70)

    # Load data with slang terms
    data_result = load_and_prepare_slang_data(max_samples_per_class=200)
    if data_result is None:
        print("❌ Failed to load data. Stopping analysis.")
        return {}

    df_data, unique_slang_terms, slang_terms_counts = data_result

    # All 4 models
    models = {
        "BART-large-MNLI": "facebook/bart-large-mnli",
        "BERT-base-MNLI": "textattack/bert-base-uncased-MNLI",
        "RoBERTa-large-MNLI": "roberta-large-mnli",
        "DeBERTa-v3-large-MNLI": "MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli"
    }

    results = {}

    for model_name, model_id in models.items():
        print(f"\n{'='*70}")
        print(f"🔍 Analyzing: {model_name}")
        print(f"{'='*70}")

        try:
            # Initialize analyzer
            analyzer = SlangAttentionAnalyzer(model_name, model_id)

            # Extract slang-focused attention patterns
            attention_data, slang_stats = analyzer.extract_slang_attention_patterns(
                df_data, max_samples=300
            )

            if len(attention_data) == 0:
                print(f"❌ No attention data extracted for {model_name}")
                continue

            # Identify slang specialist heads
            head_specializations = analyzer.identify_slang_specialist_heads(
                attention_data, slang_stats
            )

            if not head_specializations:
                print(f"❌ No slang specialists identified for {model_name}")
                continue

            # Create visualizations
            viz_filename = analyzer.visualize_slang_specialists(
                head_specializations, slang_terms_counts
            )

            # Create detailed summary
            summary_df = analyzer.create_slang_analysis_summary(
                head_specializations, slang_terms_counts
            )

            # Store results
            results[model_name] = {
                'head_specializations': head_specializations,
                'summary_df': summary_df,
                'viz_filename': viz_filename,
                'attention_data': attention_data,
                'slang_stats': slang_stats,
                'num_layers': analyzer.num_layers,
                'num_heads': analyzer.num_heads,
                'processed_examples': len(attention_data)
            }

        except Exception as e:
            print(f"❌ Error analyzing {model_name}: {str(e)}")
            import traceback
            traceback.print_exc()
            continue

    # Create cross-model comparative analysis
    if results:
        print(f"\n{'='*80}")
        print(f"📊 CROSS-MODEL SLANG SPECIALIST COMPARISON")
        print(f"{'='*80}")

        # Compare top specialists across models
        comparison_data = []

        for model_name, result in results.items():
            if result['summary_df'] is not None:
                df = result['summary_df']
                top_specialist = df.iloc[0] if len(df) > 0 else None

                if top_specialist is not None:
                    comparison_data.append({
                        'model': model_name,
                        'best_layer': int(top_specialist['layer']),
                        'best_head': int(top_specialist['head']),
                        'best_score': top_specialist['specialization_score'],
                        'consistency': top_specialist['consistency'],
                        'unique_terms': top_specialist['unique_slang_terms'],
                        'total_specialists': len(df[df['specialization_score'] > 0]),
                        'avg_specialization': df['specialization_score'].mean(),
                        'num_layers': result['num_layers'],
                        'num_heads': result['num_heads']
                    })

        if comparison_data:
            comparison_df = pd.DataFrame(comparison_data)

            # Display comparison table
            print(f"\n📋 MODEL COMPARISON TABLE:")
            print(f"{'Model':<25} {'Best Head':<12} {'Best Score':<12} {'Specialists':<12} {'Avg Score':<12}")
            print("-" * 80)

            for _, row in comparison_df.iterrows():
                model = row['model']
                best_head = f"L{row['best_layer']}-H{row['best_head']}"
                best_score = row['best_score']
                specialists = row['total_specialists']
                avg_score = row['avg_specialization']
                print(f"{model:<25} {best_head:<12} {best_score:<12.4f} {specialists:<12} {avg_score:<12.4f}")

            # Create cross-model visualization
            fig, axes = plt.subplots(2, 2, figsize=(16, 12))
            fig.suptitle('Cross-Model Slang Specialist Comparison', fontsize=16, fontweight='bold')

            # 1. Best specialist scores
            ax1 = axes[0, 0]
            ax1.bar(comparison_df['model'], comparison_df['best_score'], alpha=0.7, color='darkblue')
            ax1.set_title('Best Slang Specialist Score by Model')
            ax1.set_ylabel('Specialization Score')
            plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

            # 2. Number of specialist heads
            ax2 = axes[0, 1]
            ax2.bar(comparison_df['model'], comparison_df['total_specialists'], alpha=0.7, color='darkgreen')
            ax2.set_title('Number of Specialist Heads by Model')
            ax2.set_ylabel('Number of Specialists')
            plt.setp(ax2.get_xticklabels(), rotation=45, ha='right')

            # 3. Average specialization scores
            ax3 = axes[1, 0]
            ax3.bar(comparison_df['model'], comparison_df['avg_specialization'], alpha=0.7, color='darkorange')
            ax3.set_title('Average Specialization Score by Model')
            ax3.set_ylabel('Average Score')
            plt.setp(ax3.get_xticklabels(), rotation=45, ha='right')

            # 4. Best layer distribution
            ax4 = axes[1, 1]
            layer_ratios = comparison_df['best_layer'] / comparison_df['num_layers']
            ax4.bar(comparison_df['model'], layer_ratios, alpha=0.7, color='purple')
            ax4.set_title('Best Specialist Layer Position (Normalized)')
            ax4.set_ylabel('Layer Position Ratio')
            plt.setp(ax4.get_xticklabels(), rotation=45, ha='right')

            plt.tight_layout()

            # Save comparison
            comparison_filename = '/content/slang_specialist_model_comparison.png'
            plt.savefig(comparison_filename, dpi=300, bbox_inches='tight')
            plt.show()

            # Save detailed results
            comparison_df.to_csv('/content/slang_specialist_comparison_summary.csv', index=False)

            for model_name, result in results.items():
                if result['summary_df'] is not None:
                    result['summary_df'].to_csv(
                        f'/content/slang_specialists_{model_name.replace("-", "_").replace("/", "_")}.csv',
                        index=False
                    )

            # Analyze slang term patterns across models
            print(f"\n🔍 SLANG TERM ANALYSIS ACROSS MODELS:")

            # Find which slang terms are most challenging/interesting
            term_analysis = {}
            for model_name, result in results.items():
                for data in result['attention_data']:
                    if data['has_slang_term'] and data['slang_term']:
                        term = data['slang_term']
                        if term not in term_analysis:
                            term_analysis[term] = {'models': {}, 'total_focus': 0, 'count': 0}

                        if model_name not in term_analysis[term]['models']:
                            term_analysis[term]['models'][model_name] = []

                        term_analysis[term]['models'][model_name].append(data['slang_focus_score'])
                        term_analysis[term]['total_focus'] += data['slang_focus_score']
                        term_analysis[term]['count'] += 1

            # Find most and least attended slang terms
            term_avg_scores = {}
            for term, data in term_analysis.items():
                if data['count'] >= 4:  # At least 4 examples across models
                    term_avg_scores[term] = data['total_focus'] / data['count']

            if term_avg_scores:
                sorted_terms = sorted(term_avg_scores.items(), key=lambda x: x[1], reverse=True)

                print(f"\n🎯 MOST ATTENDED SLANG TERMS (across all models):")
                for i, (term, score) in enumerate(sorted_terms[:10], 1):
                    print(f"   {i:2d}. '{term}': {score:.4f}")

                print(f"\n🔍 LEAST ATTENDED SLANG TERMS (across all models):")
                for i, (term, score) in enumerate(sorted_terms[-10:], 1):
                    print(f"   {i:2d}. '{term}': {score:.4f}")

            print(f"\n💾 Results saved:")
            print(f"   📁 slang_specialist_comparison_summary.csv")
            print(f"   📁 Individual specialist CSV files for each model")
            print(f"   📁 Visualization files for each model")
            print(f"   📁 slang_specialist_model_comparison.png")

        else:
            print("❌ No successful model analyses for comparison")

    else:
        print("❌ No models were successfully analyzed")

    print(f"\n🎉 Slang Specialist Analysis Complete!")
    print(f"\n📝 SUMMARY:")
    print(f"   Analyzed {len(results)} models successfully")
    print(f"   Identified slang specialist attention heads in each model")
    print(f"   Analyzed attention patterns for {len(unique_slang_terms)} unique slang terms")
    print(f"   Most common slang terms: {list(slang_terms_counts.head().index)}")

    return results

# =============================================================================
# ADDITIONAL ANALYSIS FUNCTIONS
# =============================================================================

def analyze_specific_slang_term(results, target_term, max_examples=5):
    """Analyze how different models process a specific slang term"""
    print(f"\n🎯 ANALYZING SPECIFIC SLANG TERM: '{target_term}'")
    print("=" * 60)

    for model_name, result in results.items():
        print(f"\n📊 {model_name}:")

        # Find examples with this slang term
        term_examples = []
        for data in result['attention_data']:
            if data['slang_term'] == target_term and data['has_slang_term']:
                term_examples.append(data)

        if not term_examples:
            print(f"   No examples found for '{target_term}'")
            continue

        # Sort by slang focus score
        term_examples.sort(key=lambda x: x['slang_focus_score'], reverse=True)

        print(f"   Found {len(term_examples)} examples")
        print(f"   Average slang focus score: {np.mean([ex['slang_focus_score'] for ex in term_examples]):.4f}")

        # Show top attention heads for this term
        head_scores = {}
        for ex in term_examples:
            key = (ex['layer'], ex['head'])
            if key not in head_scores:
                head_scores[key] = []
            head_scores[key].append(ex['slang_focus_score'])

        # Find best heads for this term
        head_averages = {k: np.mean(v) for k, v in head_scores.items() if len(v) >= 2}
        if head_averages:
            best_heads = sorted(head_averages.items(), key=lambda x: x[1], reverse=True)[:5]
            print(f"   Top heads for '{target_term}':")
            for (layer, head), score in best_heads:
                print(f"     Layer {layer}, Head {head}: {score:.4f}")

        # Show example sentences
        print(f"   Example sentences (top {min(max_examples, len(term_examples))}):")
        for i, ex in enumerate(term_examples[:max_examples], 1):
            sentence = ex['sentence'][:100] + "..." if len(ex['sentence']) > 100 else ex['sentence']
            score = ex['slang_focus_score']
            print(f"     {i}. (Score: {score:.3f}) {sentence}")

def create_attention_visualization_for_term(results, target_term, model_name):
    """Create detailed attention visualization for a specific slang term"""
    if model_name not in results:
        print(f"Model {model_name} not found in results")
        return

    result = results[model_name]

    # Find examples with this term
    term_examples = [data for data in result['attention_data']
                    if data['slang_term'] == target_term and data['has_slang_term']]

    if not term_examples:
        print(f"No examples found for '{target_term}' in {model_name}")
        return

    # Take the best example (highest slang focus score)
    best_example = max(term_examples, key=lambda x: x['slang_focus_score'])

    # Create attention heatmap
    fig, ax = plt.subplots(figsize=(12, 10))

    tokens = best_example['tokens']
    attention_matrix = best_example['attention_matrix']
    slang_positions = best_example['slang_positions']

    # Truncate for visualization
    max_tokens = 20
    if len(tokens) > max_tokens:
        tokens = tokens[:max_tokens]
        attention_matrix = attention_matrix[:max_tokens, :max_tokens]
        slang_positions = [pos for pos in slang_positions if pos < max_tokens]

    # Create heatmap
    im = ax.imshow(attention_matrix, cmap='Blues', aspect='auto')

    # Set labels
    ax.set_xticks(range(len(tokens)))
    ax.set_yticks(range(len(tokens)))
    ax.set_xticklabels(tokens, rotation=45, ha='right')
    ax.set_yticklabels(tokens)

    # Highlight slang token positions
    for pos in slang_positions:
        ax.axhline(y=pos, color='red', linewidth=3, alpha=0.7)
        ax.axvline(x=pos, color='red', linewidth=3, alpha=0.7)

    # Add colorbar
    plt.colorbar(im, ax=ax, label='Attention Weight')

    # Set title
    layer = best_example['layer']
    head = best_example['head']
    score = best_example['slang_focus_score']
    ax.set_title(f'{model_name}: "{target_term}" - Layer {layer}, Head {head}\n'
                f'Slang Focus Score: {score:.4f}\n'
                f'Sentence: {best_example["sentence"][:80]}...', fontsize=12)

    plt.tight_layout()

    # Save
    filename = f'/content/attention_viz_{target_term}_{model_name.replace("-", "_").replace("/", "_")}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()

    return filename

# =============================================================================
# RUN ANALYSIS
# =============================================================================

if __name__ == "__main__":
    # Run the complete slang specialist analysis
    slang_results = run_slang_specialist_analysis()

    # Example of additional analyses you can run:
    #
    # # Analyze a specific slang term across all models
    # if slang_results:
    #     analyze_specific_slang_term(slang_results, "cool")
    #
    #     # Create detailed attention visualization for a term
    #     create_attention_visualization_for_term(slang_results, "cool", "BERT-base-MNLI")
```
